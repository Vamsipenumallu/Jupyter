```python
def fact(a):
    x=1
    for i in range (1,a+1):
        x=x*i
    print(x)
```

```python
fact(5)
```

```python
def armstrong(n):
    ns=str(n)
    n1=len(ns)
    summ=0
    for i in ns:
        summ=summ+(int(i)**n1)
    if(summ==n):
        return True
    return False
```

```python
armstrong(369)
```
